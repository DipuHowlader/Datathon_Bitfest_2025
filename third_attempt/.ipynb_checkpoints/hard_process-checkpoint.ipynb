```python
# educational_results
def parse_educational_result(result_str):
    result_str = result_str.lower()  
    result_str = re.sub(r'[^0-9.%cgpa]', '', result_str) 
    if '%' in result_str:
        try:
            return float(result_str.replace('%', '')) / 100
        except ValueError:
            return None
    elif 'cgpa' in result_str or 'gpa' in result_str:
        try:
            return float(re.search(r'\d+(\.\d+)?', result_str).group())
        except AttributeError:
            return None
    else:
        try:
            return float(result_str)
        except ValueError:
            return None

def clean_educational_result():
    train_data['educational_results'].fillna("unknown", inplace=True)
    train_data['parsed_educational_result'] = train_data['educational_results'].apply(parse_educational_result)

    # Step 3: Feature Engineering
    train_data['is_percentage'] = train_data['educational_results'].str.contains('%', case=False, na=False).astype(int)
    train_data['is_gpa'] = train_data['educational_results'].str.contains('gpa|cgpa', case=False, na=False).astype(int)


clean_educational_result()
```

```python
field_mapping = {
    r'\bcs\b': 'computer science',
    r'\bcomp sci\b': 'computer science',
    r'\bit\b': 'information technology',
    r'\binfo tech\b': 'information technology',
    r'\bengg\b': 'engineering',
    r'\beng\b': 'engineering',
    r'\bmech eng\b': 'mechanical engineering',
    r'\belec eng\b': 'electrical engineering',
    r'\bmath\b': 'mathematics',
    r'\bstat\b': 'mathematics',
    r'\bbiz\b': 'business',
    r'\bmgt\b': 'management',
    r'\bfin\b': 'finance',
    r'\bcomm eng\b': 'communication engineering'
    }

def clean_and_standardize_field(field):
    field = field.lower() 
    field = re.sub(r'[^a-z\s]', '', field) 
    field = re.sub(r'\s+', ' ', field).strip()  
    for pattern, replacement in field_mapping.items():
        field = re.sub(pattern, replacement, field)
    return field
    

def clean_major_field_of_studies():
    train_data['major_field_of_studies'].fillna("unknown", inplace=True)
    train_data['major_field_cleaned'] = train_data['major_field_of_studies'].apply(clean_and_standardize_field)
    common_fields = ['computer science', 'engineering', 'business', 'management', 'mathematics', 'finance']
    for field in common_fields:
        train_data[f'is_{field.replace(" ", "_")}'] = train_data['major_field_cleaned'].apply(
            lambda x: int(field in x)
        )

    
    for pattern, replacement in field_mapping.items():
        field = re.sub(pattern, replacement, field)  # Replace abbreviations
    return field
clean_major_field_of_studies()
```

```python
import re
import ast

def preprocess_company_name(name):
    """
    Clean and normalize individual company names.
    """
    if not isinstance(name, str) or name.lower() in ['n/a', 'unknown', 'company name', 'not specified']:
        return "unknown"
    name = name.lower()
    name = re.sub(r'[^a-z\s&\d]', '', name)  # Remove special characters except '&'
    name = re.sub(r'\s+', ' ', name).strip()  # Normalize whitespace
    return name

def preprocess_company_names(company_names):
    """
    Preprocess the professional_company_names column.
    """
    try:
        # Handle None values and invalid entries
        if company_names is None or not isinstance(company_names, str):
            return []
        # Convert string representation of lists into Python lists
        company_list = ast.literal_eval(company_names) if isinstance(company_names, str) else []
        # Clean individual company names
        company_list = [preprocess_company_name(name) for name in company_list if isinstance(name, str)]
        return list(set(company_list))  # Remove duplicates
    except (ValueError, SyntaxError):
        # Return an empty list for invalid entries
        return []

def preprocess_and_engineer_features(train_data):
    """
    Apply preprocessing and feature engineering to the professional_company_names column.
    """
    # Fill missing values with a placeholder
    train_data['professional_company_names'].fillna("not specified", inplace=True)
    
    # Preprocess company names
    train_data['processed_company_names'] = train_data['professional_company_names'].apply(preprocess_company_names)
    
    # Extract and normalize company names into a single string for encoding
    train_data['company_names_str'] = train_data['processed_company_names'].apply(lambda x: ' '.join(x))
    
    # Extract number of unique company names per entry
    train_data['num_companies'] = train_data['processed_company_names'].apply(len)
    
    # Feature Engineering: Identify keywords in company names
    keywords = ['tech', 'consulting', 'software', 'finance', 'marketing', 'retail', 'manufacturing', 'university']
    for keyword in keywords:
        train_data[f'is_{keyword}'] = train_data['processed_company_names'].apply(
            lambda x: int(any(keyword in name for name in x))
        )
    
    # Feature Engineering: Extract the length of the longest company name
    train_data['longest_company_name_length'] = train_data['processed_company_names'].apply(
        lambda x: max((len(name) for name in x), default=0)
    )
    
    # Feature Engineering: Count occurrences of the word "unknown"
    train_data['num_unknown_companies'] = train_data['processed_company_names'].apply(
        lambda x: sum(1 for name in x if name == "unknown")
    )
    
    return train_data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_features(train_data)
```

```python
import re
from urllib.parse import urlparse

def preprocess_company_url(url):
    """
    Normalize and extract domain from the company URL.
    """
    if not isinstance(url, str) or url.lower() in ['n/a', 'unknown', 'not specified']:
        return "unknown"
    
    # Extract domain name from URL
    try:
        domain = urlparse(url).netloc  # Extract domain
        domain = domain.replace("www.", "").strip().lower()  # Remove "www." and normalize case
        domain = re.sub(r'[^a-z0-9\-\.]', '', domain)  # Remove special characters
        return domain.split('.')[0]  # Keep the main part of the domain (e.g., 'google' from 'google.com')
    except Exception:
        return "invalid"

def preprocess_company_urls(train_data):
    """
    Preprocess and engineer features from the company_urls column.
    """
    # Fill missing values
    train_data['company_urls'].fillna("not specified", inplace=True)
    
    # Preprocess URLs to extract domains
    train_data['processed_company_urls'] = train_data['company_urls'].apply(preprocess_company_url)
    
    # Feature Engineering: Count unique URLs per entry
    train_data['num_unique_urls'] = train_data['processed_company_urls'].apply(
        lambda x: len(set(x.split())) if isinstance(x, str) else 0
    )
    
    # Feature Engineering: Check for presence of common domain types
    domain_types = ['.com', '.org', '.edu', '.gov', '.net']
    for domain in domain_types:
        train_data[f'contains_{domain[1:]}'] = train_data['company_urls'].apply(
            lambda x: int(domain in x.lower()) if isinstance(x, str) else 0
        )
    
    # Feature Engineering: Identify invalid or unknown URLs
    train_data['is_invalid_url'] = train_data['processed_company_urls'].apply(
        lambda x: 1 if x in ['unknown', 'invalid'] else 0
    )
    
    return train_data

# Apply preprocessing and feature engineering
train_data = preprocess_company_urls(train_data)

# Display the first few rows of the processed data
display(train_data[['company_urls', 'processed_company_urls', 'num_unique_urls', 'is_invalid_url'] + 
           [f'contains_{domain[1:]}' for domain in ['.com', '.org', '.edu', '.gov', '.net']]].head(70))
```

```python
from datetime import datetime
import pandas as pd

def preprocess_and_engineer_start_dates(data):
    """
    Preprocess and engineer features for the start_dates column.
    """
    def parse_date_entry(entry):
        """
        Parse individual date entries from the list.
        Replace invalid entries with None.
        """
        try:
            return pd.to_datetime(entry, errors='coerce').date()
        except:
            return None
    
    def process_start_dates_column(dates):
        """
        Process the entire start_dates entry (which might be a list).
        Select the earliest valid date from the list.
        """
        if not isinstance(dates, str):
            return None
        try:
            dates_list = eval(dates)  # Convert string representation of list to actual list
            parsed_dates = [parse_date_entry(date.strip()) for date in dates_list if date]
            valid_dates = list(filter(None, parsed_dates))  # Exclude None values
            return min(valid_dates) if valid_dates else None  # Get the earliest valid date
        except:
            return None

    # Preprocess and parse the 'start_dates' column
    data['processed_start_dates'] = data['start_dates'].apply(process_start_dates_column)
    
    # Feature Engineering
    data['start_year'] = data['processed_start_dates'].apply(lambda x: x.year if pd.notnull(x) else None)
    data['start_month'] = data['processed_start_dates'].apply(lambda x: x.month if pd.notnull(x) else None)
    data['days_since_start'] = data['processed_start_dates'].apply(
        lambda x: (pd.Timestamp(datetime.now().date()) - pd.Timestamp(x)).days if pd.notnull(x) else -1
    )
    data['start_in_future'] = data['days_since_start'].apply(lambda x: 1 if x < 0 else 0)

    return data

# Apply preprocessing and feature engineering
if 'start_dates' in train_data.columns:
    train_data = preprocess_and_engineer_start_dates(train_data)
```

```python
from datetime import datetime
import pandas as pd

def preprocess_and_engineer_end_dates(data):
    """
    Preprocess and engineer features for the end_dates column.
    """
    def parse_date_entry(entry):
        """
        Parse individual date entries from the list.
        Replace 'Till Date' or 'Current' with today's date.
        """
        if entry in ['Till Date', 'Current']:
            return datetime.now().date()
        try:
            return pd.to_datetime(entry, errors='coerce').date()
        except:
            return None
    
    def process_end_dates_column(dates):
        """
        Process the entire end_dates entry (which might be a list).
        Select the most recent valid date from the list.
        """
        if not isinstance(dates, str):
            return None
        try:
            dates_list = eval(dates)  # Convert string representation of list to actual list
            parsed_dates = [parse_date_entry(date.strip()) for date in dates_list if date]
            valid_dates = list(filter(None, parsed_dates))  # Exclude None values
            return max(valid_dates) if valid_dates else None  # Get the most recent valid date
        except:
            return None

    # Preprocess and parse the 'end_dates' column
    data['processed_end_dates'] = data['end_dates'].apply(process_end_dates_column)
    
    # Feature Engineering
    data['end_year'] = data['processed_end_dates'].apply(lambda x: x.year if pd.notnull(x) else None)
    data['end_month'] = data['processed_end_dates'].apply(lambda x: x.month if pd.notnull(x) else None)
    data['days_until_end'] = data['processed_end_dates'].apply(
        lambda x: (pd.Timestamp(x) - pd.Timestamp(datetime.now().date())).days if pd.notnull(x) else -1
    )
    data['end_in_past'] = data['days_until_end'].apply(lambda x: 1 if x < 0 else 0)

    return data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_end_dates(train_data)



display(train_data[['processed_end_dates', 'end_dates','end_year', 'end_month', 'days_until_end', 'end_in_past' ]])
```

```python
import re

def preprocess_and_engineer_related_skills(data):
    """
    Preprocess and engineer features for the related_skils_in_job column.
    """
    def clean_skills(text):
        """
        Clean and normalize the skills list.
        """
        if not isinstance(text, str):
            return "unknown"
        text = text.lower()
        text = re.sub(r'[^a-z0-9,\s]', '', text)  # Remove special characters except commas
        text = re.sub(r'\s+', ' ', text).strip()  # Normalize whitespace
        return text
    
    def count_skills(text):
        """
        Count the number of skills listed.
        """
        if text == "unknown":
            return 0
        return len([skill.strip() for skill in text.split(',') if skill.strip()])
    
    def skill_keyword_match(text, keywords):
        """
        Check for the presence of specific keywords in the skills list.
        """
        return int(any(keyword in text for keyword in keywords))

    # Clean the related_skils_in_job column
    data['related_skils_in_job_cleaned'] = data['related_skils_in_job'].apply(clean_skills)
    
    # Feature Engineering: Count the number of skills
    data['num_skills'] = data['related_skils_in_job_cleaned'].apply(count_skills)
    
    # Feature Engineering: Check for the presence of specific skills
    skill_keywords = ['python', 'java', 'excel', 'sql', 'leadership', 'data']
    for keyword in skill_keywords:
        data[f'contains_{keyword}'] = data['related_skils_in_job_cleaned'].apply(
            lambda x: skill_keyword_match(x, [keyword])
        )
    
    return data
skill_keywords = ['python', 'java', 'excel', 'sql', 'leadership', 'data']
# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_related_skills(train_data)
display(train_data[['related_skils_in_job_cleaned', 'num_skills'] + [f'contains_{key}' for key in skill_keywords]])
```

```python
import re

def preprocess_and_engineer_locations(data):
    """
    Preprocess and engineer features for the locations column.
    """
    def clean_locations(text):
        """
        Clean and normalize the location list.
        """
        if not isinstance(text, str):
            return "unknown"
        text = text.lower()
        text = re.sub(r'[^a-z0-9,\s]', '', text)  # Remove special characters except commas
        text = re.sub(r'\s+', ' ', text).strip()  # Normalize whitespace
        return text
    
    def count_locations(text):
        """
        Count the number of locations listed.
        """
        if text == "unknown":
            return 0
        return len([loc.strip() for loc in text.split(',') if loc.strip()])
    
    def location_keyword_match(text, keywords):
        """
        Check for the presence of specific location keywords.
        """
        return int(any(keyword in text for keyword in keywords))
    
    def classify_location(text):
        """
        Classify the type of location: remote, international, or domestic.
        """
        if "remote" in text:
            return "remote"
        elif any(keyword in text for keyword in ['usa', 'uk', 'canada', 'india', 'bangladesh']):
            return "domestic"
        else:
            return "international"

    # Clean the locations column
    data['locations_cleaned'] = data['locations'].apply(clean_locations)
    
    # Feature Engineering: Count the number of locations
    data['num_locations'] = data['locations_cleaned'].apply(count_locations)
    
    # Feature Engineering: Check for specific location keywords
    location_keywords = ['remote', 'usa', 'india', 'bangladesh', 'uk', 'canada']
    for keyword in location_keywords:
        data[f'contains_{keyword}'] = data['locations_cleaned'].apply(
            lambda x: location_keyword_match(x, [keyword])
        )
    
    # Feature Engineering: Classify location type
    data['location_type'] = data['locations_cleaned'].apply(classify_location)
    
    return data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_locations(train_data)

# Display the processed DataFrame
display(train_data[['locations_cleaned', 'num_locations'] + [f'contains_{key}' for key in ['remote', 'usa', 'india', 'bangladesh', 'uk', 'canada']]].head(50))
```

```python
def preprocess_and_engineer_certification_providers(data):
    """
    Preprocess and engineer features for the certification_providers column,
    including an 'other' category for uncommon providers.
    """
    def clean_certification_providers(providers):
        """
        Clean and normalize the list of certification providers.
        """
        if pd.isna(providers):
            return ["unknown"]
        try:
            # Convert string representation of list to an actual list
            providers_list = eval(providers) if isinstance(providers, str) else providers
            cleaned = [
                re.sub(r'[^a-z0-9\s]', '', provider.strip().lower()) for provider in providers_list if provider
            ]
            return list(set(cleaned))  # Remove duplicates
        except:
            return ["unknown"]

    def categorize_providers(providers, common_providers):
        """
        Separate common and uncommon providers.
        """
        if not isinstance(providers, list):
            return {"common": [], "other": []}
        common = [provider for provider in providers if provider in common_providers]
        other = [provider for provider in providers if provider not in common_providers]
        return {"common": common, "other": other}

    # Clean the certification_providers column
    data['certification_providers_cleaned'] = data['certification_providers'].apply(clean_certification_providers)
    
    # Define common providers
    common_providers = ['coursera', 'linkedin learning', 'udemy', 'google', 'microsoft', 'aws']
    
    # Categorize providers into common and other
    categorized_providers = data['certification_providers_cleaned'].apply(
        lambda x: categorize_providers(x, common_providers)
    )
    data['common_providers'] = categorized_providers.apply(lambda x: x['common'])
    data['other_providers'] = categorized_providers.apply(lambda x: x['other'])
    
    # Feature Engineering: Count the number of providers in each category
    data['num_common_providers'] = data['common_providers'].apply(len)
    data['num_other_providers'] = data['other_providers'].apply(len)
    
    return data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_certification_providers(train_data)

# Display the processed DataFrame
display(train_data[['certification_providers_cleaned','num_certification_providers', 'other_providers'] + [f'is_{key.replace(" ", "_")}' for key in  ['coursera', 'linkedin learning', 'udemy', 'google', 'microsoft', 'aws']]].head(30))
```

```python
from urllib.parse import urlparse

def preprocess_and_engineer_online_links(data):
    """
    Preprocess and engineer features for the online_links column.
    """
    def clean_links(links):
        """
        Clean and normalize the list of links.
        """
        if pd.isna(links) or links == "[None]":
            return ["unknown"]
        try:
            links_list = eval(links) if isinstance(links, str) else links
            cleaned = [
                urlparse(link).netloc.replace("www.", "").strip().lower() if link else "unknown"
                for link in links_list
            ]
            return list(set(filter(None, cleaned)))  # Remove duplicates and empty entries
        except:
            return ["unknown"]

    def count_links(links):
        """
        Count the number of valid links in the list.
        """
        return len([link for link in links if link != "unknown"]) if isinstance(links, list) else 0

    def contains_invalid_links(links):
        """
        Check if the list contains invalid or unknown links.
        """
        return int(any(link == "unknown" for link in links)) if isinstance(links, list) else 1

    # Clean the online_links column
    data['online_links_cleaned'] = data['online_links'].apply(clean_links)
    
    # Feature Engineering: Count the number of links
    data['num_links'] = data['online_links_cleaned'].apply(count_links)
    
    # Feature Engineering: Check for invalid links
    data['has_invalid_links'] = data['online_links_cleaned'].apply(contains_invalid_links)
    
    return data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_online_links(train_data)

# Display a sample of the processed data
display(train_data[['online_links_cleaned','num_links', 'has_invalid_links']])
```

```python
from datetime import datetime
import pandas as pd

def preprocess_and_engineer_issue_dates(data):
    """
    Preprocess and engineer features for the issue_dates column.
    """
    def parse_date_entry(entry):
        """
        Parse individual date entries. Return None for invalid entries.
        """
        try:
            return pd.to_datetime(entry, errors='coerce').date()
        except:
            return None

    def process_issue_dates_column(dates):
        """
        Process the entire issue_dates entry (which might be a list).
        Select the earliest valid date from the list.
        """
        if pd.isna(dates):
            return None
        try:
            dates_list = eval(dates) if isinstance(dates, str) else dates
            parsed_dates = [parse_date_entry(date.strip()) for date in dates_list if date]
            valid_dates = list(filter(None, parsed_dates))  # Exclude None values
            return min(valid_dates) if valid_dates else None  # Get the earliest valid date
        except:
            return None

    # Preprocess and parse the 'issue_dates' column
    data['processed_issue_dates'] = data['issue_dates'].apply(process_issue_dates_column)
    
    # Feature Engineering
    data['issue_year'] = data['processed_issue_dates'].apply(lambda x: x.year if pd.notnull(x) else None)
    data['issue_month'] = data['processed_issue_dates'].apply(lambda x: x.month if pd.notnull(x) else None)
    data['days_since_issue'] = data['processed_issue_dates'].apply(
        lambda x: (pd.Timestamp(datetime.now().date()) - pd.Timestamp(x)).days if pd.notnull(x) else -1
    )
    
    return data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_issue_dates(train_data)

# Display the processed DataFrame
display(train_data[['issue_year','issue_month', 'days_since_issue']].head(30))
```

```python
from datetime import datetime
import pandas as pd

def preprocess_and_engineer_expiry_dates(data):
    """
    Preprocess and engineer features for the expiry_dates column.
    """
    def parse_date_entry(entry):
        """
        Parse individual date entries and convert to a consistent format.
        """
        try:
            return pd.to_datetime(entry, errors='coerce').date()
        except:
            return None
    
    def process_expiry_dates_column(dates):
        """
        Process the expiry_dates entry (which might be a list).
        Select the most recent valid date.
        """
        if pd.isna(dates):
            return None
        try:
            dates_list = eval(dates) if isinstance(dates, str) else dates
            parsed_dates = [parse_date_entry(date.strip()) for date in dates_list if date]
            valid_dates = list(filter(None, parsed_dates))  # Exclude None values
            return max(valid_dates) if valid_dates else None  # Get the most recent valid date
        except:
            return None

    # Preprocess and parse the 'expiry_dates' column
    data['processed_expiry_dates'] = data['expiry_dates'].apply(process_expiry_dates_column)
    
    # Feature Engineering
    data['expiry_year'] = data['processed_expiry_dates'].apply(lambda x: x.year if pd.notnull(x) else None)
    data['expiry_month'] = data['processed_expiry_dates'].apply(lambda x: x.month if pd.notnull(x) else None)
    data['days_until_expiry'] = data['processed_expiry_dates'].apply(
        lambda x: (pd.Timestamp(x) - pd.Timestamp(datetime.now().date())).days if pd.notnull(x) else -1
    )
    data['is_expired'] = data['days_until_expiry'].apply(lambda x: 1 if x < 0 else 0)

    return data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_expiry_dates(train_data)
```

```python
import re

def preprocess_and_engineer_education(data):
    """
    Preprocess and engineer features for the educationaL_requirements column.
    """
    def clean_education(text):
        """
        Normalize and clean education requirements text.
        """
        if not isinstance(text, str):
            return "unknown"
        text = text.lower()
        text = re.sub(r'[^a-z\s&()]', '', text)  # Remove special characters
        text = re.sub(r'\s+', ' ', text).strip()  # Normalize whitespace
        return text

    def extract_degrees(text):
        """
        Extract degree types from the education requirements.
        """
        degrees = ['bachelor', 'master', 'phd', 'diploma', 'honors']
        return [deg for deg in degrees if deg in text]

    def extract_fields(text):
        """
        Extract fields of study from the education requirements.
        """
        fields = ['computer science', 'engineering', 'mechanical engineering', 'civil engineering']
        return [field for field in fields if field in text]
    
    def map_education_level(degrees):
        """
        Map degree types to an ordinal scale.
        """
        level_mapping = {'diploma': 1, 'bachelor': 2, 'master': 3, 'phd': 4}
        levels = [level_mapping[deg] for deg in degrees if deg in level_mapping]
        return max(levels) if levels else 0  # Return highest level or 0 if none

    # Clean the educationaL_requirements column
    data['education_cleaned'] = data['educationaL_requirements'].apply(clean_education)
    
    # Extract degrees and fields
    data['degrees_extracted'] = data['education_cleaned'].apply(extract_degrees)
    data['fields_extracted'] = data['education_cleaned'].apply(extract_fields)
    
    # Feature Engineering: Map education levels
    data['education_level'] = data['degrees_extracted'].apply(map_education_level)
    
    # Feature Engineering: Create binary flags for degrees and fields
    degrees = ['bachelor', 'master', 'phd', 'diploma', 'honors']
    for degree in degrees:
        data[f'is_{degree}'] = data['degrees_extracted'].apply(lambda x: int(degree in x))
    
    fields = ['computer science', 'engineering', 'mechanical engineering', 'civil engineering']
    for field in fields:
        data[f'is_{field.replace(" ", "_")}'] = data['fields_extracted'].apply(lambda x: int(field in x))
    
    return data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_education(train_data)

```

```python
import re

def preprocess_and_engineer_experience_requirement(data):
    """
    Preprocess and engineer features for the experiencere_requirement column.
    """
    def clean_experience_text(text):
        """
        Clean and normalize the experience requirement text.
        """
        if pd.isna(text):
            return "unknown"
        text = text.lower()
        text = re.sub(r'\(s\)', '', text)  # Remove plural indicators
        text = re.sub(r'[^a-z0-9\s]', '', text)  # Remove special characters
        text = re.sub(r'\s+', ' ', text).strip()  # Normalize whitespace
        return text
    
    def extract_min_experience(text):
        """
        Extract the minimum experience from the text.
        """
        match = re.search(r'(\d+)', text)
        return int(match.group(1)) if match else 0
    
    def extract_max_experience(text):
        """
        Extract the maximum experience for ranges (e.g., '1 to 3 years').
        """
        match = re.search(r'(\d+)\s+to\s+(\d+)', text)
        return int(match.group(2)) if match else None
    
    def categorize_experience(min_exp):
        """
        Categorize experience into common ranges.
        """
        if min_exp < 2:
            return "<2 years"
        elif 2 <= min_exp <= 5:
            return "2-5 years"
        else:
            return ">5 years"

    # Clean the experiencere_requirement column
    data['experience_requirement_cleaned'] = data['experiencere_requirement'].apply(clean_experience_text)
    
    # Extract minimum and maximum experience
    data['min_experience'] = data['experience_requirement_cleaned'].apply(extract_min_experience)
    data['max_experience'] = data['experience_requirement_cleaned'].apply(extract_max_experience)
    
    # Categorize experience
    data['experience_category'] = data['min_experience'].apply(categorize_experience)
    
    # Feature Engineering: Flags for common experience ranges
    data['is_less_than_2_years'] = data['min_experience'].apply(lambda x: int(x < 2))
    data['is_between_2_and_5_years'] = data['min_experience'].apply(lambda x: int(2 <= x <= 5))
    data['is_more_than_5_years'] = data['min_experience'].apply(lambda x: int(x > 5))
    
    return data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_experience_requirement(train_data)

# Display the processed DataFrame
```

```python
def preprocess_and_engineer_age_requirement(data):
    """
    Preprocess and engineer features for the age_requirement column.
    """
    def parse_age_requirement(requirement):
        """
        Extract min and max age from the requirement text.
        """
        if pd.isna(requirement):
            return {"min_age": None, "max_age": None}
        
        requirement = requirement.lower()
        
        # Extract ranges like "22 to 30 years"
        match_range = re.search(r'(\d+)\s*to\s*(\d+)', requirement)
        if match_range:
            return {"min_age": int(match_range.group(1)), "max_age": int(match_range.group(2))}
        
        # Extract "Above 25 years" or "Under 18 years"
        match_above = re.search(r'above\s*(\d+)', requirement)
        match_under = re.search(r'under\s*(\d+)', requirement)
        if match_above:
            return {"min_age": int(match_above.group(1)), "max_age": None}
        if match_under:
            return {"min_age": None, "max_age": int(match_under.group(1))}
        
        # No valid age found
        return {"min_age": None, "max_age": None}

    # Apply parsing logic
    parsed_ages = data['age_requirement'].apply(parse_age_requirement)
    
    # Extract min and max age into separate columns
    data['min_age'] = parsed_ages.apply(lambda x: x['min_age'])
    data['max_age'] = parsed_ages.apply(lambda x: x['max_age'])
    
    # Feature Engineering: Flag if age is specified
    data['is_age_specified'] = data['min_age'].notnull() | data['max_age'].notnull()
    
    # Replace nulls with a placeholder for text representation
    data['age_requirement_cleaned'] = data['age_requirement'].fillna("unknown")
    
    return data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_age_requirement(train_data)
```

```python
def preprocess_and_engineer_responsibilities(data):
    """
    Preprocess and engineer features for the responsibilities.1 column.
    """
    def clean_responsibilities(text):
        """
        Clean and normalize responsibilities text.
        """
        if pd.isna(text):
            return "unknown"
        text = text.lower()
        text = re.sub(r'[^a-z0-9\s\n]', '', text)  # Remove special characters
        text = re.sub(r'\s+', ' ', text).strip()  # Normalize whitespace
        return text

    def count_responsibilities(text):
        """
        Count the number of responsibilities (lines) in the text.
        """
        if text == "unknown":
            return 0
        return len(text.split('\n'))

    def keyword_match(text, keywords):
        """
        Check for the presence of specific keywords in the responsibilities.
        """
        if text == "unknown":
            return 0
        return int(any(keyword in text for keyword in keywords))

    # Clean the responsibilities.1 column
    data['responsibilities_cleaned'] = data['responsibilities.1'].apply(clean_responsibilities)
    
    # Feature Engineering: Count the number of responsibilities
    data['num_responsibilities'] = data['responsibilities.1'].apply(
        lambda x: len(x.split('\n')) if isinstance(x, str) else 0
    )
    
    # Feature Engineering: Check for specific keywords
    responsibility_keywords = ['machine learning', 'data analysis', 'troubleshooting', 'leadership', 'quality']
    for keyword in responsibility_keywords:
        data[f'contains_{keyword.replace(" ", "_")}'] = data['responsibilities_cleaned'].apply(
            lambda x: keyword_match(x, [keyword])
        )
    
    return data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_responsibilities(train_data)
```

```python
import re

def preprocess_and_engineer_skills_required(data):
    """
    Preprocess and engineer features for the skills_required column.
    """
    def clean_skills(text):
        """
        Clean and normalize the list of skills.
        """
        if pd.isna(text):
            return ["unknown"]
        # Split skills using common delimiters
        skills = re.split(r'[,\n;]', text.lower())
        # Normalize and clean individual skills
        cleaned_skills = [re.sub(r'[^a-z0-9\s]', '', skill.strip()) for skill in skills if skill.strip()]
        return list(set(cleaned_skills))  # Remove duplicates

    def count_skills(skills):
        """
        Count the number of unique skills listed.
        """
        return len(skills) if (isinstance(skills, list) and len(skills) > 0 and skills[0] != 'unknown' and skills[0] != 'na')  else 0

    def skill_keyword_match(skills, keyword):
        """
        Check for the presence of specific skill keywords in the list.
        """
        if not isinstance(skills, list):
            return 0
        return int(keyword in skills)

    # Clean the skills_required column
    data['skills_required_cleaned'] = data['skills_required'].apply(clean_skills)
    
    # Feature Engineering: Count the number of unique skills
    data['num_skills_required'] = data['skills_required_cleaned'].apply(count_skills)
    
    # Feature Engineering: Check for specific skill keywords
    skill_keywords = ['python', 'java', 'communication', 'marketing', 'excel', 'data', 'leadership']
    for keyword in skill_keywords:
        data[f'contains_{keyword}'] = data['skills_required_cleaned'].apply(
            lambda x: skill_keyword_match(x, keyword)
        )
    
    return data

# Apply preprocessing and feature engineering
train_data = preprocess_and_engineer_skills_required(train_data)
```
